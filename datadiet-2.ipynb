```python
import pandas as pd
import numpy as np

df_1 = pd.read_csv('/Users/iwatashun/Desktop/datadiet/datadiet_database - health_record_d1.csv')
df_d2 = pd.read_csv('/Users/iwatashun/Desktop/datadiet/datadiet_database - ideal_nutrition_d2-2.csv')
df_3 = pd.read_csv('/Users/iwatashun/Desktop/datadiet/datadiet_database - menu_nutrition_d3.csv')
```

## df_1 to df_2

```python
#df_1 to df_2
rep = pd.DataFrame()
rep['resident_id'] = df_1['resident_id']
cals = []
for i in df_1['bmr']:
    mul = df_1[df_1['bmr'] == i]['activity_level:\n1 = sedentary\n2 = lightly active\n3 = moderately active\n4 = very active\n5 = extra active'].iloc[0]
    if mul == 1:
        cals.append(round(i*1.2, 2))
    elif mul == 2:
        cals.append(round(i*1.375, 2))
    elif mul == 3:
        cals.append(round(i*1.55, 2))
    elif mul == 4:
        cals.append(round(i*1.725, 2))
    else:
        cals.append(round(i*1.9, 2))
        
rep['calories'] = cals
rep['restrictions'] = df_1['restrictions']
water = []
for i in range(len(df_1['weight (kg)'])):
    w = df_1['weight (kg)'][i] * 30
    if df_1['dehydration'][i] == True:
        w += 500
    water.append(w)
rep['water (mL)'] = water


s_f = []
for i in range(len(rep['calories'])):
    if df_1['dehydration'][i] == True or df_1['type_2_diabetes'][i] == True or df_1['cardiovascular_disease'][i] == True:
        s_f.append(f"< {round(rep['calories'][i] * 0.07 * 1/9, 2)}")
    else:
        s_f.append(f"< {round(rep['calories'][i] * 0.1 * 1/9,2)}")
rep['saturated_fat (g)'] = s_f
t_f = []
for i in range(len(rep['calories'])):
    if df_1['dehydration'][i] == True or df_1['hypertension'][i] == True or df_1['cardiovascular_disease'][i] == True:
        t_f.append(f" < {round(rep['calories'][i] * 0.01 * 1/9, 2)}")
    else:
        t_f.append(f"< {round(rep['calories'][i] * 0.008 * 1/9, 2)}")
rep['trans_fat (g)'] = t_f

cho = []
for i in range(len(rep['calories'])):
    if df_1['dehydration'][i] == True or df_1['hypertension'][i] == True or df_1['cardiovascular_disease'][i] == True:
        cho.append("< 200")
    else:
        cho.append("< 300")
rep['cholesterol (mg)'] = cho

sod = []
for i in range(len(rep['calories'])):
    if df_1['hypertension'][i] == True or df_1['cardiovascular_disease'][i] == True:
        sod.append("< 1500")
    else:
        sod.append("< 2000")
rep['sodium (mg)'] = sod

t_c = []
for i in range(len(rep['calories'])):
    if df_1['type_2_diabetes'][i] == True:
        t_c.append(f"< {round(rep['calories'][i] * 0.45 /4)}")
    else:
        t_c.append(f"{round(rep['calories'][i] * 0.45 /4)}-{round(rep['calories'][i] * 0.65 /4)}")
rep['total_carbs (g)'] = t_c

d_f =[]

for i in range(len(rep['calories'])):
    if df_1['gender'][i] == 'M':
        d_f.append(max([round(rep['calories'][i]*14/1000,2),30]))
    else:
        d_f.append(max([round(rep['calories'][i]*14/1000,2),21]))
rep['dietary_fiber (g)'] = d_f

t_s = []
for i in range(len(rep['calories'])):
    if df_1['type_2_diabetes'][i] == True or df_1['hypertension'][i] == True or df_1['cardiovascular_disease'][i] == True:
        t_s.append(f"< {round(rep['calories'][i] * 0.05/4,2)}")
    else:
        t_s.append(f"< {round(rep['calories'][i] * 0.1/4,2)}")
rep['total_sugars (g)'] = t_s

pro = []
for i in range(len(rep['calories'])):
    if df_1['osteoporosis'][i] == True or df_1['respiratory_disease'][i] == True:
        pro.append(f"> {round(df_1['weight (kg)'][i] * 1.2,1)}")
    else:
        pro.append(f"> {df_1['weight (kg)'][i]}")
rep['protein (g)'] = pro

pot = []
for i in range(len(rep['calories'])):
    if df_1['gender'][i] == 'M':
        if df_1['dehydration'][i] == True or  df_1['type_2_diabetes'][i] == True or df_1['hypertension'][i] == True:
            pot.append(4400)
        else:
            pot.append(3400)
    else:
        if df_1['dehydration'][i] == True or  df_1['type_2_diabetes'][i] == True or df_1['hypertension'][i] == True:
            pot.append(3600)
        else:
            pot.append(2600)
rep['potassium (mg)'] = pot

c = []
for i in range(len(rep['calories'])):
    if df_1['osteoporosis'][i] == True or df_1['hypertension'][i] == True:
        c.append("> 1200")
    else:
        c.append("> 1000")
rep['calcium (mg)'] = c

iron = []
for i in range(len(rep['calories'])):
    if df_1['gastrointestinal_disorder'][i] == True or df_1['cardiovascular_disease'][i] == True:
        iron.append(15)
    else:
        iron.append(8)
rep['iron (mg)'] = iron

vit = []
for i in range(len(rep['calories'])):
    if df_1['osteoporosis'][i] == True or df_1['gastrointestinal_disorder'][i] == True or df_1['respiratory_disease'][i] == True:
        vit.append("> 25")
    else:
        vit.append("> 20")
rep['vitamin_d (mcg)'] = vit
```

```python
df_2 = rep.copy()
df_2 = df_2.head(10)
```

## changing the datatype (getting rid of <,>, and -)

```python

variables = df_2.columns
less = ['saturated_fat (g)','trans_fat (g)','cholesterol (mg)', 'sodium (mg)','total_sugars (g)']
more = ['vitamin_d (mcg)','calcium (mg)','protein (g)']
for i in variables:
    if i in ['resident_id','calories','water (mL)','dietary_fiber (g)','potassium (mg)','iron (mg)']:
        df_2[i] = pd.to_numeric(df_2[i])
    else:
        df_2[i] = df_2[i].astype(str)
        if i in less or i in more:
            num = []
            for j in range(len(df_2[i])):
                num.append(float(df_2[i][j].split()[1]))
            df_2[i] = num
        elif i == 'total_carbs (g)':
            num = []
            for j in range(len(df_2[i])):
                if len(df_2['total_carbs (g)'][j]) == 5:
                    num.append(float(df_2['total_carbs (g)'][j].split()[1]))
                else:
                    a_1 = float(df_2['total_carbs (g)'][j].split('-')[0])
                    a_2 = float(df_2['total_carbs (g)'][j].split('-')[1])
                    ave = (a_1 + a_2) / 2
                    num.append(ave)
            df_2[i] = num    
```

## importing the dictionary from step_2

```python
day = {}
day = pd.read_pickle('/Users/iwatashun/Desktop/datadiet/step2.pickle')
meal_type = list(day.keys())
```

```python
#creating all possible combination of 
meals = pd.DataFrame()
for j in df_3.columns:
    sub = []
    if j in ['Menu Items', 'type', 'restrictions']:
        pass
    else:
        for k in day.keys():
            for i in day[k]:
                a = df_3['Menu Items'].isin(list(i))
                choice = df_3[a].fillna(0)
                sub.append(sum(choice[j]))
        meals[j] = sub
types = []
for i in day.keys():
    types += [i] * len(day[i])
meals['type'] = types    

bf_opt = meals[meals['type'] == 'breakfast']
lun_opt = meals[meals['type'] == 'lunch']
din_opt = meals[meals['type'] == 'dinner']
```

```python
final_options = pd.DataFrame() #df that combines all the nutrition for the combination including label in dictionay 'day'
for i in range(len(bf_opt.index)):
    for j in range(len(lun_opt.index)):
        for k in range(len(din_opt.index)):
            option = pd.DataFrame()
            option = pd.concat([bf_opt.iloc[[i]],lun_opt.iloc[[j]],din_opt.iloc[[k]]],axis = 0)
            row = {}
            for l in option.columns:
                if l == 'type':
                    pass
                else:
                    row[l] = sum(option[l])
            row['combo'] = str(i) + str(j) + str(k)
            info = pd.DataFrame.from_dict(row, orient = 'index')
            info = info.transpose()
            final_options = pd.concat([final_options, info], axis = 0)
            
final_options.reset_index(drop=True, inplace=True) 
for i in final_options.columns:
    if i == 'combo':
        pass
    else:
        final_options[i] = pd.to_numeric(final_options[i])
```

```python
test = final_options.copy()#copy of final menu options
no_need = ['resident_id', 'restrictions']
cols = list(df_2.columns)
cols.remove('resident_id')
cols.remove('restrictions')
comp = df_2[cols]#nutrition of each samples
test = test[cols]#copy of final menu options
dfs = {} #dic that stores df of all the difference from samples' standard
for i in range(len(df_2)):
    test = final_options.copy()
    test = test[cols]
    for j in range(len(test)):
        test.loc[j,:] = test.loc[j,:] - comp.iloc[i,:]
    test['water (mL)'].values[:] = 0
    dfs[i] = test
```

```python
final = pd.DataFrame()
for j in range(len(df_2['resident_id'])):
    a = dfs[j][more]
    penalties = []
    for i in more:
        penalties.append(abs(a[[i]]).max())
    a = a[a[more]>0].fillna(-1)
    for i in range(len(penalties)):
        a[[more[i]]] = a[[more[i]]].replace(-1,penalties[i])
    for i in more:
        a[i] = (a[i]/comp.loc[0,i])**2

    b = dfs[0][less]
    penalties = []
    for i in less:
        penalties.append(abs(b[[i]]).max())
    b = b[b[less]<0].fillna(1)
    for i in range(len(penalties)):
        b[[less[i]]] = b[[less[i]]].replace(1,penalties[i])
    for i in less:
        b[i] = (b[i]/comp.loc[0,i])**2

    r = dfs[0].columns[~ dfs[0].columns.isin(more)]
    r = r[~ r.isin(less)]
    c = dfs[0][r]
    for i in r:
        c[i] = (c[i]/comp.loc[0,i])**2

    scores = pd.concat([a,b,c], axis = 1)
    scores['scores'] = scores.sum(axis=1)

    scores.nsmallest(5, ['scores']).index
    top_options = final_options.iloc[list(scores.nsmallest(5, ['scores']).index),:]
    sub = pd.DataFrame()
    a = 1
    for i in top_options['combo']:
        sub['resident_id'] = [df_2['resident_id'][j]]*3
        sub['option'] = [a]*3
        main = []
        side = []
        m_type = []
        for k in range(len(i)):
            main.append(day[meal_type[k]][int(i[k])][0])
            side.append(day[meal_type[k]][int(i[k])][1])
            m_type.append(meal_type[k])
        sub['main'] = main
        sub['side'] = side
        sub['meal_type'] = m_type
        final = pd.concat([final, sub.reset_index()],axis = 0)
        a += 1
 
```
